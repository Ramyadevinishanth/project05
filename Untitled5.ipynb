```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

IMG_SIZE = 224
BATCH_SIZE = 32

# -----------------------------
#  TRAIN AUGMENTATION
# -----------------------------
train_transform = transforms.Compose([
    # Zoom + Crop + Resize (best for aerial images)
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.7, 1.0)),

    # Rotation
    transforms.RandomRotation(30),

    # Horizontal Flip
    transforms.RandomHorizontalFlip(p=0.5),

    # Brightness / Contrast / Saturation / Hue
    transforms.ColorJitter(
        brightness=0.3,
        contrast=0.3,
        saturation=0.3,
        hue=0.05
    ),

    # Shift + Shear
    transforms.RandomAffine(
        degrees=0,
        translate=(0.2, 0.2),
        shear=20
    ),

    transforms.ToTensor(),

    # Normalization
    transforms.Normalize([0.5, 0.5, 0.5],
                         [0.5, 0.5, 0.5])
])

# -----------------------------
#  VALID / TEST TRANSFORM
# -----------------------------
test_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5],
                         [0.5, 0.5, 0.5])
])

# -----------------------------
#  DATASETS
# -----------------------------
train_dataset = datasets.ImageFolder(
    '/content/drive/MyDrive/ariealfile/train',
    transform=train_transform
)

val_dataset = datasets.ImageFolder(
    '/content/drive/MyDrive/ariealfile/valid',
    transform=test_transform
)

test_dataset = datasets.ImageFolder(
    '/content/drive/MyDrive/ariealfile/test',
    transform=test_transform
)

# -----------------------------
#  DATALOADERS
# -----------------------------
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        # Convolution Layers
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding=1, stride=1)
        self.bn1 = nn.BatchNorm2d(32)  # Batch Normalization
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1, stride=1)
        self.bn2 = nn.BatchNorm2d(64)  # Batch Normalization

        # Fully Connected Layers
        self.gap = nn.AdaptiveAvgPool2d((1, 1))
        self.fc1 = nn.Linear(64 , 128)  # assuming input images 224x224
        self.dropout = nn.Dropout(0.5)          # Dropout for regularization
        self.fc2 = nn.Linear(128, 2)            # Binary classification: Bird / Drone

    def forward(self, x):
        # Conv Layer 1 --> ReLU --> BatchNorm --> Pooling
        x = self.pool(torch.relu(self.bn1(self.conv1(x))))

        # Conv Layer 2 --> ReLU --> BatchNorm --> Pooling
        x = self.pool(torch.relu(self.bn2(self.conv2(x))))

        # Flatten

        x = self.gap(x)
        x = torch.flatten(x, 1)

        # Fully connected layer 1 --> ReLU --> Dropout
        x = self.dropout(torch.relu(self.fc1(x)))

        # Fully connected layer 2 --> Output
        x = self.fc2(x)
        return x

# Instantiate model
model = CNN()

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import torch
import torch.nn as nn
from torchvision import models
from torchvision.models import ResNet50_Weights, MobileNet_V2_Weights, EfficientNet_B0_Weights
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

```

```python
resnet50 = models.resnet50(weights=ResNet50_Weights.DEFAULT)
for param in resnet50.parameters():
    param.requires_grad = False  # freeze layers
resnet50.fc = nn.Linear(resnet50.fc.in_features, 2)  # 2 classes: Bird/Drone
resnet50 = resnet50.to(device)
```

```python
mobilenet = models.mobilenet_v2(weights=MobileNet_V2_Weights.DEFAULT)
for param in mobilenet.parameters():
    param.requires_grad = False
mobilenet.classifier[1] = nn.Linear(mobilenet.classifier[1].in_features, 2)
mobilenet = mobilenet.to(device)
```

```python
efficientnet = models.efficientnet_b0(weights=EfficientNet_B0_Weights.DEFAULT)
for param in efficientnet.parameters():
    param.requires_grad = False
efficientnet.classifier[1] = nn.Linear(efficientnet.classifier[1].in_features, 2)
efficientnet = efficientnet.to(device)
```

```python
import torch
import os

class EarlyStopping:
    def __init__(self, patience=5, checkpoint_path="best_model.pth", verbose=False):
        self.patience = patience
        self.best_loss = float('inf')
        self.counter = 0
        self.early_stop = False
        self.checkpoint_path = checkpoint_path
        self.verbose = verbose

    def __call__(self, val_loss, model):
        if val_loss < self.best_loss:
            self.best_loss = val_loss
            self.counter = 0

            # Save best model
            torch.save(model.state_dict(), self.checkpoint_path)
            if self.verbose:
                print("Validation loss improved  model saved.")

        else:
            self.counter += 1
            if self.verbose:
                print(f"No improvement ({self.counter}/{self.patience})")

            if self.counter >= self.patience:
                self.early_stop = True

    def load_best_model(self, model):
        if os.path.exists(self.checkpoint_path):
            model.load_state_dict(torch.load(self.checkpoint_path))
            print("Loaded best saved model.")
        return model
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import torch

def train_model(model, train_loader, val_loader, criterion, optimizer, device, epochs=10):
    early_stopping = EarlyStopping(patience=5, verbose=True)

    for epoch in range(epochs):
        # --- Training ---
        model.train()
        train_loss = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        # --- Validation ---
        model.eval()
        val_loss = 0
        all_preds, all_labels = [], []
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                preds = torch.argmax(outputs, dim=1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        # --- Metrics ---
        acc = accuracy_score(all_labels, all_preds)
        prec = precision_score(all_labels, all_preds )
        rec = recall_score(all_labels, all_preds)
        f1 = f1_score(all_labels, all_preds)

        print(f"Epoch [{epoch+1}/{epochs}] "
              f"Train Loss: {train_loss/len(train_loader):.4f} "
              f"Val Loss: {val_loss/len(val_loader):.4f} "
              f"Acc: {acc:.4f} Prec: {prec:.4f} Rec: {rec:.4f} F1: {f1:.4f}")

        # --- Early Stopping ---
        early_stopping(val_loss/len(val_loader), model)
        if early_stopping.early_stop:
            print("Early stopping triggered!")
            break

    # Load the best model
    model = early_stopping.load_best_model(model)
    return model
```

```python
import torch

# ============================================
# DEVICE SETUP
# ============================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using:", device)

# ============================================
# MOVE MODELS TO GPU
# ============================================
model = model.to(device)
resnet50 = resnet50.to(device)
mobilenet = mobilenet.to(device)
efficientnet = efficientnet.to(device)

# ============================================
# LOSS FUNCTION
# ============================================
criterion = torch.nn.CrossEntropyLoss().to(device)

# ============================================
# OPTIMIZERS
# (Only update final layer for transfer learning models)
# ============================================
optimizer_cnn = torch.optim.Adam(model.parameters(), lr=0.001)
optimizer_resnet = torch.optim.Adam(resnet50.fc.parameters(), lr=0.0001)
optimizer_mobilenet = torch.optim.Adam(mobilenet.classifier[1].parameters(), lr=0.0001)
optimizer_efficientnet = torch.optim.Adam(efficientnet.classifier[1].parameters(), lr=0.0001)

# ============================================
# TRAIN DIFFERENT MODELS
# ============================================
best_cnn = train_model(
    model, train_loader, val_loader,
    criterion, optimizer_cnn, device, epochs=10
)

best_resnet = train_model(
    resnet50, train_loader, val_loader,
    criterion, optimizer_resnet, device, epochs=10
)

best_mobilenet = train_model(
    mobilenet, train_loader, val_loader,
    criterion, optimizer_mobilenet, device, epochs=10
)

best_efficientnet = train_model(
    efficientnet, train_loader, val_loader,
    criterion, optimizer_efficientnet, device, epochs=10
)

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import torch

# Function to evaluate a model
def evaluate_model(model, test_loader, device):
    model.eval()
    all_preds, all_labels = [], []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            preds = torch.argmax(outputs, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    acc = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds, average='binary')
    rec = recall_score(all_labels, all_preds, average='binary')
    f1 = f1_score(all_labels, all_preds, average='binary')

    return acc, prec, rec, f1


# Evaluate all models
cnn_metrics = evaluate_model(best_cnn, test_loader, device)
resnet_metrics = evaluate_model(best_resnet, test_loader, device)
mobilenet_metrics = evaluate_model(best_mobilenet, test_loader, device)
efficientnet_metrics = evaluate_model(best_efficientnet, test_loader, device)

# Print comparison table
print("\n=== Model Performance Comparison ===\n")
print("| Model         | Accuracy | Precision | Recall | F1-score |")
print("|---------------|----------|-----------|--------|----------|")
print(f"| Custom CNN    | {cnn_metrics[0]:.3f}   | {cnn_metrics[1]:.3f}    | {cnn_metrics[2]:.3f}  | {cnn_metrics[3]:.3f} |")
print(f"| ResNet50      | {resnet_metrics[0]:.3f}   | {resnet_metrics[1]:.3f}    | {resnet_metrics[2]:.3f}  | {resnet_metrics[3]:.3f} |")
print(f"| MobileNet     | {mobilenet_metrics[0]:.3f}   | {mobilenet_metrics[1]:.3f}    | {mobilenet_metrics[2]:.3f}  | {mobilenet_metrics[3]:.3f} |")
print(f"| EfficientNet  | {efficientnet_metrics[0]:.3f}   | {efficientnet_metrics[1]:.3f}    | {efficientnet_metrics[2]:.3f}  | {efficientnet_metrics[3]:.3f} |")

```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import torch

def plot_confusion_matrix(model, test_loader, device, title):
    model.eval()
    all_preds, all_labels = [], []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            preds = torch.argmax(outputs, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)

    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Bird","Drone"],
                yticklabels=["Bird","Drone"])
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Classification Report
    print(f"\n===== Classification Report: {title} =====\n")
    print(classification_report(all_labels, all_preds, target_names=["Bird","Drone"]))


# Run for all models
plot_confusion_matrix(best_cnn, test_loader, device, "Custom CNN")
plot_confusion_matrix(best_resnet, test_loader, device, "ResNet50")
plot_confusion_matrix(best_mobilenet, test_loader, device, "MobileNet")
plot_confusion_matrix(best_efficientnet, test_loader, device, "EfficientNet")
```

```python
num_epochs = 3
train_losses = []
val_losses = []
train_acc = []
val_acc = []

for epoch in range(num_epochs):
    train_loss_epoch = 0
    train_correct = 0
    total_train = 0

    best_cnn.train()
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = best_cnn(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss_epoch += loss.item()
        _, predicted = torch.max(outputs, 1)
        train_correct += (predicted == labels).sum().item()
        total_train += labels.size(0)

    # Validation
    val_loss_epoch = 0
    val_correct = 0
    total_val = 0
    best_cnn.eval()
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = best_cnn(inputs)
            loss = criterion(outputs, labels)
            val_loss_epoch += loss.item()
            _, predicted = torch.max(outputs, 1)
            val_correct += (predicted == labels).sum().item()
            total_val += labels.size(0)

    # Store per-epoch metrics
    train_losses.append(train_loss_epoch / len(train_loader))
    val_losses.append(val_loss_epoch / len(val_loader))
    train_acc.append(train_correct / total_train)
    val_acc.append(val_correct / total_val)

history_cnn = {
    'train_loss': train_losses,
    'val_loss': val_losses,
    'train_acc': train_acc,
    'val_acc': val_acc
}
```

```python
def plot_training_curves(history, title):
    plt.figure(figsize=(12,5))

    # Loss curve
    plt.subplot(1,2,1)
    plt.plot(history['train_loss'], label="Train Loss")
    plt.plot(history['val_loss'], label="Validation Loss")
    plt.title(f"{title} - Loss Curve")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()

    # Accuracy curve
    plt.subplot(1,2,2)
    plt.plot(history['train_acc'], label="Train Accuracy")
    plt.plot(history['val_acc'], label="Validation Accuracy")
    plt.title(f"{title} - Accuracy Curve")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.show()
```
